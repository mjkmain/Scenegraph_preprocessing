- Scene graph json to csv

```python
import json
import pandas as pd

with open("../data/v1/scene_graph/scene_graph.json", 'r') as f:
    sg = json.load(f)

arr = []

for s in sg:
    obj_dict = dict()
    for o in s['OBJECTS']:
        obj_dict[o['OBJECT_ID']] = o['NAME']


    for o in s['OBJECTS']:
        
        critera_obj = obj_dict[o['OBJECT_ID']]
        for r in o["RELATIONS"]:
            temp_dict = dict()
            temp_dict['ID'] = s['Scene_Graph_ID']
            temp_dict['obj1'] = critera_obj
            temp_dict['rel'] = r['NAME']
            temp_dict['obj2'] = obj_dict[r['OBJECT']]

            arr.append(temp_dict)
sg_df = pd.DataFrame(arr)
sg_df
```

```python
sg_df.to_csv("../data/v1/scene_graph/scene_graph.csv", index=False)
```
