```python
import json
import pandas as pd

with open("../data/v1/scene_graph/scene_graph.json", 'r') as f:
    sg = json.load(f)

arr = []
bounding_box_dict = dict()

for s in sg:
    obj_dict = dict()
    bbox_dict = dict()
    for o in s['OBJECTS']:
        obj_dict[o['OBJECT_ID']] = o['NAME']
        bbox_dict[obj_dict[o['OBJECT_ID']]] = f"{o['H']}, {o['W']}, {o['X']}, {o['Y']}"

    bounding_box_dict[s['Scene_Graph_ID']] = bbox_dict

    for o in s['OBJECTS']:
        
        critera_obj = obj_dict[o['OBJECT_ID']]
        for r in o["RELATIONS"]:
            temp_dict = dict()
            temp_dict['ID'] = s['Scene_Graph_ID']
            temp_dict['img_name'] = s['IMAGE_NAME']
            temp_dict['obj1'] = critera_obj
            temp_dict['rel'] = r['NAME']
            temp_dict['obj2'] = obj_dict[r['OBJECT']]


            arr.append(temp_dict)

sg_df = pd.DataFrame(arr)
sg_df
```

```python
with open("../data/v1/scene_graph/bbox.json", 'w') as f:
    json.dump(bounding_box_dict, f, ensure_ascii=False, indent=4)
```
